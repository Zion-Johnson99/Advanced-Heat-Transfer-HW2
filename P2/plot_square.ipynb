```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# 1. 创建一个图形(figure)和一个坐标轴(axes)
fig, ax = plt.subplots()

# 2. 定义一个方块
# patches.Rectangle((x, y), width, height, ...)
# (0, 0) 是方块左下角的坐标
# 'facecolor' 是填充颜色
square_fill = patches.Rectangle(
    (0, 0),                   # 方块的左下角坐标
    1,                        # 宽度
    1,                        # 高度
    facecolor='#a2d08a',      # 填充颜色 (一个类似的淡绿色)
    edgecolor='black'         # 边框颜色
)

# 3. 将方块添加到坐标轴上
ax.add_patch(square_fill)

# 4. 设置坐标轴的范围，以便能完整显示方块
ax.set_xlim(-0.5, 1.5)
ax.set_ylim(-0.5, 1.5)

# 5. 设置坐标轴的宽高比为'equal'，确保方块是正方形
ax.set_aspect('equal')

# 6. 关闭坐标轴的刻度和边框
ax.axis('off')

# 7. 显示图形
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# 关键！告诉Jupyter在单元格中显示图形
%matplotlib inline  

# 1. 创建图形和坐标轴
fig, ax = plt.subplots(figsize=(6, 6))

# 2. 定义方块（无填充）
square = patches.Rectangle(
    (0, 0),                   # 左下角坐标
    1,                        # 宽度
    1,                        # 高度
    facecolor='#a2d08a',         # 填充颜色
    edgecolor='black',        # 边框颜色
    linewidth=2               
)

# 3. 将方块添加到坐标轴
ax.add_patch(square)

# 4. 添加内部的文字和点 (已根据新图像调整位置)
ax.text(0.5, 0.55, r'$T(x,y)$', ha='center', va='center', fontsize=18)
ax.plot(0.5, 0.25, 'ko', markersize=6) # 'ko' = black (k) circle (o)
ax.text(0.5, 0.35, r'$T_i$', ha='center', va='center', fontsize=16)

# 5. 添加边界条件（箭头和标签）
common_arrow_props = dict(
    arrowstyle='->',    # 箭头样式
    lw=1.5,             # 箭头线条宽度
    color='black'
)

common_text_props = dict(
    ha='center',        # 水平居中
    va='center',        # 垂直居中
    fontsize=14
)

# 上边界: q''_o (根据您的新图片修改)
ax.annotate(
    r'$q^{\prime\prime}_o$',  # <--- 使用LaTeX标准双引号表示
    xy=(0.5, 1), 
    xytext=(0.5, 1.3),
    arrowprops=common_arrow_props,
    **common_text_props
)

# 左边界: q''_o (根据您的新图片修改)
ax.annotate(
    r'$q^{\prime\prime}_o$',  # <--- 使用LaTeX标准双引号表示
    xy=(0, 0.5), 
    xytext=(-0.3, 0.5),
    arrowprops=common_arrow_props,
    **common_text_props
)

# 右边界: T_o
ax.annotate(
    r'$T_o$', 
    xy=(1, 0.5), 
    xytext=(1.3, 0.5),
    arrowprops=common_arrow_props,
    **common_text_props
)

# 下边界: T_o
ax.annotate(
    r'$T_o$', 
    xy=(0.5, 0), 
    xytext=(0.5, -0.3),
    arrowprops=common_arrow_props,
    **common_text_props
)

# 6. 设置坐标轴范围并隐藏
ax.set_xlim(-0.5, 1.5)
ax.set_ylim(-0.5, 1.5)
ax.set_aspect('equal') # 保持正方形
ax.axis('off')         # 隐藏坐标轴

# 7. 显示最终图形
# plt.show() 在 %matplotlib inline 模式下是可选的，但加上也没问题
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# 关键！告诉Jupyter在单元格中显示图形
%matplotlib inline

# 1. 创建图形和坐标轴
fig, ax = plt.subplots(figsize=(18, 6)) # 调整figsize以适应更宽的图

# 2. 定义三个方块的位置和大小
square_width = 1.5
square_height = 1.5
square_spacing = 3.5 # 增加间距以容纳等号和加号

# 方块1位置 (θ(β, α))
square1_x = 0
square1_y = 0

# 方块2位置 (θ₁(β, α))
square2_x = square1_x + square_width + square_spacing
square2_y = 0

# 方块3位置 (θ₂(β, α))
square3_x = square2_x + square_width + square_spacing
square3_y = 0

# 3. 创建三个绿色方块
square1 = patches.Rectangle(
    (square1_x, square1_y),
    square_width,
    square_height,
    facecolor='#a2d08a',
    edgecolor='black',
    linewidth=2
)

square2 = patches.Rectangle(
    (square2_x, square2_y),
    square_width,
    square_height,
    facecolor='#a2d08a',
    edgecolor='black',
    linewidth=2
)

square3 = patches.Rectangle(
    (square3_x, square3_y),
    square_width,
    square_height,
    facecolor='#a2d08a',
    edgecolor='black',
    linewidth=2
)

# 4. 将方块添加到坐标轴
ax.add_patch(square1)
ax.add_patch(square2)
ax.add_patch(square3)

# 5. 添加方块中心的文字
ax.text(square1_x + square_width/2, square1_y + square_height/2, r'$\theta(\alpha, \beta)$', 
        ha='center', va='center', fontsize=16, weight='bold')
ax.text(square2_x + square_width/2, square2_y + square_height/2, r'$\theta_1(\alpha, \beta)$', 
        ha='center', va='center', fontsize=16, weight='bold')
ax.text(square3_x + square_width/2, square3_y + square_height/2, r'$\theta_2(\alpha, \beta)$', 
        ha='center', va='center', fontsize=16, weight='bold')

# 6. 添加等号和加号 (位置已调整到红框指示的区域)
# 等号位于方块1和方块2之间
ax.text(square1_x + square_width + square_spacing/2 - 0.5, square1_y + square_height/2, # 微调x坐标
        r'$=$', ha='center', va='center', fontsize=24, color='black', weight='bold')
# 加号位于方块2和方块3之间
ax.text(square2_x + square_width + square_spacing/2 - 0.5, square2_y + square_height/2, # 微调x坐标
        r'$+$', ha='center', va='center', fontsize=24, color='black', weight='bold')


# 7. 添加边界条件（箭头和标签）
common_arrow_props = dict(
    arrowstyle='->',
    lw=1.5,
    color='black'
)

common_text_props = dict(
    ha='center',
    va='center',
    fontsize=12
)

# 方块1的边界条件 (θ(α, β))
# 上边界
ax.annotate(r'$\frac{\partial\theta}{\partial \alpha}|_{\alpha=1} = 1$',
            xy=(square1_x + square_width/2, square1_y + square_height),
            xytext=(square1_x + square_width/2, square1_y + square_height + 0.5),
            arrowprops=common_arrow_props, **common_text_props)

# 左边界
ax.annotate(r'$\frac{\partial\theta}{\partial \beta}|_{\beta=0} = -1$',
            xy=(square1_x, square1_y + square_height/2),
            xytext=(square1_x - 0.8, square1_y + square_height/2),
            arrowprops=common_arrow_props, **common_text_props)

# 右边界
ax.annotate(r'$0$',
            xy=(square1_x + square_width, square1_y + square_height/2),
            xytext=(square1_x + square_width + 0.4, square1_y + square_height/2),
            arrowprops=common_arrow_props, **common_text_props)

# 下边界
ax.annotate(r'$0$',
            xy=(square1_x + square_width/2, square1_y),
            xytext=(square1_x + square_width/2, square1_y - 0.4),
            arrowprops=common_arrow_props, **common_text_props)

# 方块2的边界条件 (θ₁(α, β))
# 上边界
ax.annotate(r'$\frac{\partial\theta_1}{\partial \alpha}|_{\alpha=1} = 1$',
            xy=(square2_x + square_width/2, square2_y + square_height),
            xytext=(square2_x + square_width/2, square2_y + square_height + 0.5),
            arrowprops=common_arrow_props, **common_text_props)

# 左边界
ax.annotate(r'$\frac{\partial\theta_1}{\partial \beta}|_{\beta=0} = 0$',
            xy=(square2_x, square2_y + square_height/2),
            xytext=(square2_x - 0.8, square2_y + square_height/2),
            arrowprops=common_arrow_props, **common_text_props)

# 右边界
ax.annotate(r'$0$',
            xy=(square2_x + square_width, square2_y + square_height/2),
            xytext=(square2_x + square_width + 0.4, square2_y + square_height/2),
            arrowprops=common_arrow_props, **common_text_props)

# 下边界
ax.annotate(r'$0$',
            xy=(square2_x + square_width/2, square2_y),
            xytext=(square2_x + square_width/2, square2_y - 0.4),
            arrowprops=common_arrow_props, **common_text_props)

# 方块3的边界条件 (θ₂(α, β))
# 上边界
ax.annotate(r'$\frac{\partial\theta_2}{\partial \alpha}|_{\alpha=1} = 0$',
            xy=(square3_x + square_width/2, square3_y + square_height),
            xytext=(square3_x + square_width/2, square3_y + square_height + 0.5),
            arrowprops=common_arrow_props, **common_text_props)

# 左边界
ax.annotate(r'$\frac{\partial\theta_2}{\partial \beta}|_{\beta=0} = -1$',
            xy=(square3_x, square3_y + square_height/2),
            xytext=(square3_x - 0.8, square3_y + square_height/2),
            arrowprops=common_arrow_props, **common_text_props)

# 右边界
ax.annotate(r'$0$',
            xy=(square3_x + square_width, square3_y + square_height/2),
            xytext=(square3_x + square_width + 0.4, square3_y + square_height/2),
            arrowprops=common_arrow_props, **common_text_props)

# 下边界
ax.annotate(r'$0$',
            xy=(square3_x + square_width/2, square3_y),
            xytext=(square3_x + square_width/2, square3_y - 0.4),
            arrowprops=common_arrow_props, **common_text_props)

# 8. 设置坐标轴范围并隐藏
ax.set_xlim(-1.5, square3_x + square_width + 1.5)
ax.set_ylim(-1.2, square_height + 1.2)
ax.set_aspect('equal')
ax.axis('off')

# 9. 显示最终图形
plt.show()
```
